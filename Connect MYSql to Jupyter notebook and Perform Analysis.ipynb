### Step 2 : Import libraries 

```python
import mysql.connector
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

### Step 3: Setup Connection to connect LocalDB to Jupyter notebook 

**NOTE**: Provide your Account's Username (user) and Password ,database is the DB you want to connect ,host will always be  localhost if you want to connect locally 

```python
db_connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="admin",
    database="hr"
)
```

```python
cursor = db_connection.cursor()
```

### Step 4 : Run a Queries 

NOTE:The OutPut will be in a List Format 
    

```python
query = """
SELECT *
FROM employees;
"""

cursor.execute(query)
```

```python
results = cursor.fetchall()
results
```

```python
query = """
SELECT *
FROM employees where salary>5000;
"""

cursor.execute(query)
```

```python
results = cursor.fetchall()
results
```

### To Convert It into DataFrame

```python
df=pd.DataFrame(results,columns=['emp_id','first_name','last_name','mail_id','cell_no','hire_date','job_id', 'salary', 
                              'commission_pct', 'manager_id', 'department_id'])
```

```python
df.head()
```

```python
df.info()
```

```python
df[df['salary']>10000]
```

```python
x=df[['first_name','salary']].sort_values('salary',ascending=False)
```

```python
x['salary']=x['salary'].astype(int)
```

```python
n=x.head(5)
```

```python
plt.style.use('seaborn')
```

```python
sns.barplot(x='first_name',y='salary' ,data=n)
```
